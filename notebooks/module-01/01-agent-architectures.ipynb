## Theory: Agent Architecture Patterns

### What is an Agentic AI System?

An **agentic AI system** is an autonomous software entity that:
- Perceives its environment through sensors or data inputs
- Reasons about goals and plans actions to achieve them
- Acts upon the environment through tools, APIs, or interfaces
- Learns and adapts from feedback and experience

### Three Core Architectures

**1. Reactive Architecture**
- Simple stimulus-response mapping
- No internal state or world model
- Fast response times
- Limited to predefined behaviors
- **Use case**: Simple chatbots, alert systems

**2. Deliberative Architecture**
- Maintains internal world model
- Uses symbolic reasoning and planning
- Goal-oriented behavior
- Handles complex, multi-step tasks
- **Use case**: Strategic planning, research agents

**3. Hybrid Architecture (Recommended)**
- Combines reactive and deliberative layers
- Fast reactive responses for routine situations
- Deliberative planning for complex scenarios
- **Use case**: Production customer service, enterprise assistants

## Implementation: Reactive Agent

A reactive agent responds directly to stimuli using predefined rules.

```python
# Implementation of a simple reactive agent
class ReactiveAgent:
    """Reactive agent that responds based on predefined rules."""
    
    def __init__(self, rules: Dict[str, Callable]):
        """
        Initialize reactive agent with rules.
        
        Args:
            rules: Dictionary mapping condition names to action functions
        """
        # Store rules as condition-action pairs
        self.rules = rules
        self.response_count = 0
    
    def perceive(self, environment: Dict[str, Any]) -> Dict[str, Any]:
        """
        Observe current environment state.
        
        Args:
            environment: Current state of the environment
        
        Returns:
            Observed state
        """
        # In reactive agents, perception is direct observation
        return environment
    
    def act(self, state: Dict[str, Any]) -> Any:
        """
        React to state based on rules.
        
        Args:
            state: Current state to react to
        
        Returns:
            Action result or None if no rule matches
        """
        # Iterate through rules and execute first matching action
        for rule_name, action_func in self.rules.items():
            try:
                # Execute action if applicable
                result = action_func(state)
                if result is not None:
                    self.response_count += 1
                    return result
            except Exception as e:
                # Handle errors gracefully
                print(f"Error in rule '{rule_name}': {e}")
                continue
        
        # No rule matched - return default response
        return "I don't understand that request."

# Define rules for a simple customer service agent
def handle_greeting(state):
    """Handle greeting messages."""
    if any(word in state.get('message', '').lower() for word in ['hello', 'hi', 'hey']):
        return "Hello! How can I help you today?"
    return None

def handle_order_status(state):
    """Handle order status inquiries."""
    if 'order' in state.get('message', '').lower() and 'status' in state.get('message', '').lower():
        return "Your order #12345 is currently being processed and will ship within 2 business days."
    return None

def handle_return(state):
    """Handle return requests."""
    if 'return' in state.get('message', '').lower():
        return "You can return items within 30 days. Visit our returns portal at example.com/returns"
    return None

# Create reactive agent with rules
reactive_agent = ReactiveAgent({
    'greeting': handle_greeting,
    'order_status': handle_order_status,
    'return': handle_return
})

# Test the reactive agent
test_messages = [
    "Hello there!",
    "What's my order status?",
    "I want to return an item",
    "Tell me about your products"
]

print("Testing Reactive Agent:\n")
for msg in test_messages:
    state = {'message': msg}
    response = reactive_agent.act(state)
    print(f"User: {msg}")
    print(f"Agent: {response}\n")

print(f"Total responses: {reactive_agent.response_count}")
```

## Implementation: Deliberative Agent

A deliberative agent maintains an internal world model and plans actions.

```python
# Implementation of a deliberative agent with planning
class DeliberativeAgent:
    """Deliberative agent that maintains world model and plans actions."""
    
    def __init__(self):
        """Initialize deliberative agent with world model and goals."""
        # Internal world model stores agent's understanding of environment
        self.world_model = {}
        # Goals the agent is trying to achieve
        self.goals = []
        # Current plan being executed
        self.current_plan = []
    
    def perceive(self, observations: Dict[str, Any]):
        """
        Update internal world model based on observations.
        
        Args:
            observations: New information from environment
        """
        # Update world model with new observations
        self.world_model.update(observations)
        print(f"Updated world model: {self.world_model}")
    
    def deliberate(self) -> List[str]:
        """
        Reason about goals and create action plan.
        
        Returns:
            List of planned actions
        """
        # Simple planning: create sequence of actions to achieve goals
        plan = []
        
        for goal in self.goals:
            # Determine actions needed for this goal
            if goal == "research_topic":
                plan.extend([
                    "search_for_information",
                    "analyze_results",
                    "synthesize_findings"
                ])
            elif goal == "answer_question":
                plan.extend([
                    "understand_question",
                    "retrieve_knowledge",
                    "formulate_answer"
                ])
        
        return plan
    
    def act(self, action: str) -> str:
        """
        Execute a planned action.
        
        Args:
            action: Action to execute
        
        Returns:
            Result of action execution
        """
        # Execute action based on type
        action_results = {
            "search_for_information": "Found 10 relevant articles",
            "analyze_results": "Identified 3 key themes",
            "synthesize_findings": "Created summary report",
            "understand_question": "Parsed question intent",
            "retrieve_knowledge": "Retrieved relevant facts",
            "formulate_answer": "Generated response"
        }
        
        result = action_results.get(action, "Action completed")
        print(f"Executing: {action} -> {result}")
        return result
    
    def run(self, goal: str):
        """
        Main execution loop: perceive, deliberate, act.
        
        Args:
            goal: Goal to achieve
        """
        # Set goal
        self.goals = [goal]
        
        # Create plan
        print(f"\nGoal: {goal}")
        plan = self.deliberate()
        print(f"Plan: {plan}\n")
        
        # Execute plan
        for action in plan:
            self.act(action)

# Test deliberative agent
delib_agent = DeliberativeAgent()

# Perceive environment
delib_agent.perceive({"user_query": "Tell me about AI agents", "context": "educational"})

# Run with goal
delib_agent.run("research_topic")
```

## Implementation: Hybrid Agent

A hybrid agent combines reactive and deliberative layers for optimal performance.

```python
# Implementation of hybrid agent architecture
class HybridAgent:
    """Hybrid agent combining reactive and deliberative approaches."""
    
    def __init__(self, reactive_rules: Dict, complexity_threshold: float = 0.5):
        """
        Initialize hybrid agent.
        
        Args:
            reactive_rules: Rules for reactive layer
            complexity_threshold: Threshold for escalating to deliberative layer
        """
        # Reactive layer for fast responses
        self.reactive_agent = ReactiveAgent(reactive_rules)
        # Deliberative layer for complex reasoning
        self.deliberative_agent = DeliberativeAgent()
        # Threshold for determining complexity
        self.complexity_threshold = complexity_threshold
    
    def assess_complexity(self, state: Dict[str, Any]) -> float:
        """
        Assess complexity of the request.
        
        Args:
            state: Current state/request
        
        Returns:
            Complexity score (0.0 to 1.0)
        """
        message = state.get('message', '').lower()
        
        # Simple heuristic: count complexity indicators
        complexity_indicators = [
            'why', 'how', 'explain', 'analyze', 'compare',
            'research', 'investigate', 'complex', 'detailed'
        ]
        
        # Calculate complexity score
        indicator_count = sum(1 for indicator in complexity_indicators if indicator in message)
        complexity = min(indicator_count / 3.0, 1.0)  # Normalize to 0-1
        
        return complexity
    
    def process(self, state: Dict[str, Any]) -> str:
        """
        Process request using appropriate layer.
        
        Args:
            state: Request state
        
        Returns:
            Response from appropriate layer
        """
        # Assess complexity
        complexity = self.assess_complexity(state)
        print(f"Complexity score: {complexity:.2f}")
        
        if complexity < self.complexity_threshold:
            # Use reactive layer for simple requests (fast path)
            print("Using reactive layer (fast path)")
            return self.reactive_agent.act(state)
        else:
            # Escalate to deliberative layer for complex requests
            print("Escalating to deliberative layer (complex reasoning)")
            self.deliberative_agent.perceive(state)
            self.deliberative_agent.run("answer_question")
            return "Complex analysis complete. See detailed reasoning above."

# Create hybrid agent
hybrid_agent = HybridAgent({
    'greeting': handle_greeting,
    'order_status': handle_order_status,
    'return': handle_return
})

# Test with various complexity levels
test_cases = [
    "Hi there!",  # Simple - reactive
    "What's my order status?",  # Simple - reactive
    "Why is my order delayed and how can I prevent this in the future?",  # Complex - deliberative
    "Explain the differences between your shipping options"  # Complex - deliberative
]

print("Testing Hybrid Agent:\n")
for msg in test_cases:
    print(f"\nUser: {msg}")
    response = hybrid_agent.process({'message': msg})
    print(f"Agent: {response}")
    print("-" * 60)
```
