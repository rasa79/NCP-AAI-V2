## Theory: RAG Architecture

### What is RAG?

**Retrieval-Augmented Generation (RAG)** combines:
1. **Retrieval**: Finding relevant documents from a knowledge base
2. **Augmentation**: Adding retrieved context to the prompt
3. **Generation**: Generating responses grounded in retrieved documents

### RAG Pipeline Components

```
Documents → Chunking → Embedding → Vector Store
                                         ↓
Query → Embedding → Retrieval → Context + Query → LLM → Response
```

### Key Benefits

- **Up-to-date information**: Access current data without retraining
- **Reduced hallucinations**: Responses grounded in source documents
- **Transparency**: Can cite sources
- **Domain-specific**: Works with specialized knowledge

````python
# Simple RAG implementation
class SimpleRAG:
    """Basic RAG system implementation."""
    
    def __init__(self):
        self.documents = []
        self.embeddings = []
    
    def add_documents(self, docs: List[str]):
        """Add documents to knowledge base."""
        self.documents.extend(docs)
        # Generate embeddings (simulated)
        for doc in docs:
            embedding = self._embed(doc)
            self.embeddings.append(embedding)
    
    def _embed(self, text: str) -> np.ndarray:
        """Generate embedding for text (simulated)."""
        # Simple hash-based embedding for demo
        return np.random.rand(384)  # Simulated 384-dim embedding
    
    def retrieve(self, query: str, k: int = 3) -> List[str]:
        """Retrieve top-k relevant documents."""
        query_embedding = self._embed(query)
        
        # Calculate similarities (simulated)
        similarities = []
        for doc_embedding in self.embeddings:
            sim = np.dot(query_embedding, doc_embedding)
            similarities.append(sim)
        
        # Get top-k indices
        top_k_indices = np.argsort(similarities)[-k:][::-1]
        
        return [self.documents[i] for i in top_k_indices]
    
    def query(self, question: str) -> str:
        """Query RAG system."""
        # Retrieve relevant documents
        context_docs = self.retrieve(question, k=2)
        
        # Build context
        context = "\n".join(context_docs)
        
        # Generate response (simulated)
        response = f"Based on the context: {context[:100]}... The answer is: [Generated response]"
        
        return response

# Test RAG system
print("=== Testing Simple RAG ===")
rag = SimpleRAG()

# Add documents
docs = [
    "RAG combines retrieval with generation for better responses.",
    "Vector databases store embeddings for fast similarity search.",
    "LangChain provides tools for building RAG applications."
]
rag.add_documents(docs)

# Query
result = rag.query("What is RAG?")
print(f"\nQuery: What is RAG?")
print(f"Response: {result}")
````
