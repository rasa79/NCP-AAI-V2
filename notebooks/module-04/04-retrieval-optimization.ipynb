```python
# Hybrid search implementation
class HybridSearch:
    """Combine semantic and keyword search."""
    
    def search(self, query: str, alpha: float = 0.5):
        """Hybrid search with weighting.
        
        Args:
            query: Search query
            alpha: Weight for semantic search (0-1)
        """
        semantic_results = self._semantic_search(query)
        keyword_results = self._keyword_search(query)
        
        # Combine results
        combined = [
            alpha * s + (1-alpha) * k 
            for s, k in zip(semantic_results, keyword_results)
        ]
        return combined
    
    def _semantic_search(self, query):
        return [0.9, 0.7, 0.5]  # Simulated scores
    
    def _keyword_search(self, query):
        return [0.6, 0.8, 0.4]  # Simulated scores

print("✅ Hybrid search ready")
```
