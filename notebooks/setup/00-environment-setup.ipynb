## Implementation: Install Core Dependencies

We'll install packages in groups to better track progress and handle potential errors.

```python
# Install core LangChain packages
# LangChain provides the foundation for agent and RAG development
import subprocess
import sys

def install_package(package_name):
    """
    Install a Python package using pip.
    
    Args:
        package_name: Name of the package to install
    
    Returns:
        bool: True if installation successful, False otherwise
    """
    try:
        # Use subprocess to run pip install command
        subprocess.check_call(
            [sys.executable, "-m", "pip", "install", "-q", package_name]
        )
        print(f"✅ Successfully installed {package_name}")
        return True
    except subprocess.CalledProcessError as e:
        # Handle installation errors gracefully
        print(f"❌ Failed to install {package_name}: {e}")
        return False
    except Exception as e:
        # Catch any unexpected errors
        print(f"❌ Unexpected error installing {package_name}: {e}")
        return False

# Install core LangChain packages
print("Installing LangChain core packages...\n")
core_packages = [
    "langchain>=0.1.0",
    "langchain-community>=0.0.10",
    "langchain-core>=0.1.0",
]

for package in core_packages:
    install_package(package)
```

```python
# Install UI and API frameworks
# Gradio for interactive interfaces, LangServe for API deployment
print("\nInstalling UI and API frameworks...\n")

ui_packages = [
    "gradio>=4.0.0",
    "langserve>=0.0.30",
    "fastapi>=0.104.0",  # Required by LangServe
    "uvicorn>=0.24.0",   # ASGI server for FastAPI
]

for package in ui_packages:
    install_package(package)
```

```python
# Install vector store libraries
# These enable semantic search and retrieval for RAG systems
print("\nInstalling vector store libraries...\n")

vector_packages = [
    "faiss-cpu>=1.7.4",        # Facebook AI Similarity Search (CPU version)
    "chromadb>=0.4.0",         # Chroma vector database
    "sentence-transformers>=2.2.0",  # For generating embeddings
]

for package in vector_packages:
    install_package(package)
```

```python
# Install testing and utility packages
# Hypothesis for property-based testing, other utilities for development
print("\nInstalling testing and utility packages...\n")

utility_packages = [
    "hypothesis>=6.90.0",      # Property-based testing framework
    "pytest>=7.4.0",           # Testing framework
    "python-dotenv>=1.0.0",    # Environment variable management
    "requests>=2.31.0",        # HTTP library for API calls
    "jupyter>=1.0.0",          # Jupyter notebook support
    "ipywidgets>=8.1.0",       # Interactive widgets for notebooks
]

for package in utility_packages:
    install_package(package)
```

## Validation: Verify Installation

Let's verify that all critical packages are installed correctly and can be imported.

```python
# Validation: Import all critical packages to verify installation
import importlib

def verify_import(module_name, display_name=None):
    """
    Verify that a module can be imported successfully.
    
    Args:
        module_name: Name of the module to import
        display_name: Optional friendly name for display
    
    Returns:
        bool: True if import successful, False otherwise
    """
    display = display_name or module_name
    try:
        # Attempt to import the module
        module = importlib.import_module(module_name)
        
        # Try to get version if available
        version = getattr(module, "__version__", "unknown")
        print(f"✅ {display}: {version}")
        return True
    except ImportError as e:
        # Handle import errors - package not installed or incompatible
        print(f"❌ {display}: Import failed - {e}")
        return False
    except Exception as e:
        # Handle unexpected errors
        print(f"⚠️  {display}: Unexpected error - {e}")
        return False

print("Verifying package installations...\n")

# Dictionary of modules to verify
modules_to_verify = {
    "langchain": "LangChain",
    "langchain_community": "LangChain Community",
    "langchain_core": "LangChain Core",
    "gradio": "Gradio",
    "langserve": "LangServe",
    "fastapi": "FastAPI",
    "faiss": "FAISS",
    "chromadb": "ChromaDB",
    "sentence_transformers": "Sentence Transformers",
    "hypothesis": "Hypothesis",
    "pytest": "Pytest",
    "dotenv": "Python-dotenv",
    "requests": "Requests",
}

# Track verification results
results = {}
for module, display in modules_to_verify.items():
    results[display] = verify_import(module, display)

# Summary of verification results
print("\n" + "="*50)
successful = sum(results.values())
total = len(results)
print(f"\nVerification Summary: {successful}/{total} packages verified")

if successful == total:
    print("\n🎉 All packages installed successfully! You're ready to proceed.")
else:
    print("\n⚠️  Some packages failed verification. Please review errors above.")
    print("   Try reinstalling failed packages or check for compatibility issues.")
```

```python
# Exercise: Test LangChain basic functionality
try:
    from langchain.schema import Document
    from langchain.text_splitter import CharacterTextSplitter
    
    # Create a sample document
    sample_text = "This is a test document for the NVIDIA RAG certification course."
    doc = Document(page_content=sample_text, metadata={"source": "test"})
    
    # Test text splitting functionality
    splitter = CharacterTextSplitter(chunk_size=20, chunk_overlap=5)
    chunks = splitter.split_documents([doc])
    
    print("✅ LangChain basic functionality test passed")
    print(f"   Created {len(chunks)} chunks from sample document")
except Exception as e:
    print(f"❌ LangChain test failed: {e}")
```

```python
# Exercise: Test FAISS vector store functionality
try:
    import faiss
    import numpy as np
    
    # Create a simple FAISS index with random vectors
    dimension = 128  # Vector dimension
    n_vectors = 10   # Number of vectors
    
    # Generate random vectors for testing
    vectors = np.random.random((n_vectors, dimension)).astype('float32')
    
    # Create FAISS index and add vectors
    index = faiss.IndexFlatL2(dimension)
    index.add(vectors)
    
    # Perform a simple search
    query_vector = np.random.random((1, dimension)).astype('float32')
    distances, indices = index.search(query_vector, k=3)
    
    print("✅ FAISS vector store test passed")
    print(f"   Indexed {index.ntotal} vectors, retrieved top 3 matches")
except Exception as e:
    print(f"❌ FAISS test failed: {e}")
```

## Troubleshooting: Common Issues

### Issue 1: Package Installation Fails

**Symptoms**: `pip install` commands fail with errors

**Solutions**:
- Upgrade pip: `pip install --upgrade pip`
- Use a virtual environment to avoid conflicts
- Check internet connectivity
- Try installing packages one at a time to identify problematic packages

### Issue 2: Import Errors After Installation

**Symptoms**: Packages install successfully but cannot be imported

**Solutions**:
- Restart the Jupyter kernel
- Verify you're using the correct Python environment
- Check for package name conflicts
- Reinstall the problematic package

### Issue 3: Version Compatibility Issues

**Symptoms**: Packages install but show version conflicts

**Solutions**:
- Use the specific versions listed in requirements.txt
- Create a fresh virtual environment
- Check package documentation for compatibility matrices

### Issue 4: Memory Errors

**Symptoms**: Installation or import fails due to insufficient memory

**Solutions**:
- Close other applications to free up RAM
- Install packages sequentially rather than all at once
- Consider using lighter alternatives (e.g., faiss-cpu instead of faiss-gpu)
