## Setup: Check GPU Availability

First, let's check if an NVIDIA GPU is available. While not required for all exercises, GPU acceleration significantly improves performance.

```python
# Check for NVIDIA GPU availability
import subprocess
import sys

def check_gpu_availability():
    """
    Check if NVIDIA GPU is available and display GPU information.
    
    Returns:
        bool: True if GPU available, False otherwise
    """
    try:
        # Try to run nvidia-smi command to check GPU status
        result = subprocess.run(
            ["nvidia-smi", "--query-gpu=name,memory.total,driver_version", "--format=csv,noheader"],
            capture_output=True,
            text=True,
            check=True
        )
        
        # Parse and display GPU information
        gpu_info = result.stdout.strip().split(", ")
        if len(gpu_info) >= 3:
            print("✅ NVIDIA GPU detected:")
            print(f"   GPU: {gpu_info[0]}")
            print(f"   Memory: {gpu_info[1]}")
            print(f"   Driver: {gpu_info[2]}")
            return True
        else:
            print("⚠️  GPU detected but unable to parse information")
            return True
    except subprocess.CalledProcessError:
        # nvidia-smi command failed - no GPU or driver issues
        print("ℹ️  No NVIDIA GPU detected or driver not installed")
        print("   CPU-only mode will be used (performance may be limited)")
        return False
    except FileNotFoundError:
        # nvidia-smi command not found
        print("ℹ️  nvidia-smi not found - no NVIDIA GPU available")
        print("   CPU-only mode will be used (performance may be limited)")
        return False
    except Exception as e:
        # Unexpected error
        print(f"⚠️  Error checking GPU: {e}")
        return False

# Check GPU availability
has_gpu = check_gpu_availability()

# Provide guidance based on GPU availability
if not has_gpu:
    print("\n💡 Note: While GPU is not required for learning, some NVIDIA tools")
    print("   work best with GPU acceleration. Consider using cloud GPU instances")
    print("   (Google Colab, AWS, Azure) for optimal experience.")
```

## Implementation: Install NeMo Guardrails

NeMo Guardrails is a Python package that can be installed via pip. It's essential for implementing safety and compliance in RAG agents.

```python
# Install NeMo Guardrails package
import subprocess
import sys

def install_nemo_guardrails():
    """
    Install NVIDIA NeMo Guardrails package.
    
    Returns:
        bool: True if installation successful, False otherwise
    """
    try:
        print("Installing NeMo Guardrails...")
        # Install nemoguardrails package with quiet output
        subprocess.check_call(
            [sys.executable, "-m", "pip", "install", "-q", "nemoguardrails>=0.7.0"]
        )
        print("✅ NeMo Guardrails installed successfully")
        return True
    except subprocess.CalledProcessError as e:
        # Handle installation failure
        print(f"❌ Failed to install NeMo Guardrails: {e}")
        print("   Try installing manually: pip install nemoguardrails")
        return False
    except Exception as e:
        # Handle unexpected errors
        print(f"❌ Unexpected error: {e}")
        return False

# Install NeMo Guardrails
install_nemo_guardrails()
```

## Implementation: Configure Docker for NIM

NVIDIA NIM runs as Docker containers. Let's verify Docker is installed and configured correctly.

```python
# Check Docker installation and configuration
import subprocess

def check_docker():
    """
    Check if Docker is installed and running.
    
    Returns:
        bool: True if Docker available, False otherwise
    """
    try:
        # Check Docker version
        result = subprocess.run(
            ["docker", "--version"],
            capture_output=True,
            text=True,
            check=True
        )
        print(f"✅ Docker installed: {result.stdout.strip()}")
        
        # Check if Docker daemon is running
        result = subprocess.run(
            ["docker", "ps"],
            capture_output=True,
            text=True,
            check=True
        )
        print("✅ Docker daemon is running")
        return True
    except subprocess.CalledProcessError:
        # Docker command failed
        print("❌ Docker is installed but not running")
        print("   Start Docker daemon and try again")
        return False
    except FileNotFoundError:
        # Docker not installed
        print("❌ Docker not found")
        print("   Install Docker from: https://docs.docker.com/get-docker/")
        return False
    except Exception as e:
        # Unexpected error
        print(f"⚠️  Error checking Docker: {e}")
        return False

# Check Docker availability
docker_available = check_docker()

if docker_available:
    print("\n💡 Docker is ready for NVIDIA NIM deployment")
    print("   You'll learn to deploy NIM containers in Module 6 notebooks")
else:
    print("\n💡 Docker is required for NIM deployment")
    print("   Install Docker before proceeding to Module 6")
```

## Troubleshooting: Common NVIDIA Platform Issues

### Issue 1: GPU Not Detected

**Symptoms**: `nvidia-smi` command fails or shows no GPU

**Solutions**:
- Install NVIDIA drivers: Visit nvidia.com/drivers
- Verify GPU is properly seated in PCIe slot
- Check BIOS settings for PCIe configuration
- For cloud instances, ensure GPU instance type is selected

### Issue 2: Docker Permission Denied

**Symptoms**: Docker commands fail with permission errors

**Solutions**:
- Add user to docker group: `sudo usermod -aG docker $USER`
- Log out and back in for group changes to take effect
- Or run Docker commands with sudo (not recommended for production)

### Issue 3: NGC API Key Invalid

**Symptoms**: Cannot pull NIM containers or access NGC resources

**Solutions**:
- Verify API key is correct (no extra spaces or characters)
- Regenerate API key from NGC console
- Check API key has not expired
- Ensure NGC_API_KEY environment variable is set correctly

### Issue 4: NeMo Guardrails Import Errors

**Symptoms**: Cannot import nemoguardrails module

**Solutions**:
- Verify installation: `pip show nemoguardrails`
- Reinstall package: `pip install --upgrade nemoguardrails`
- Check Python version compatibility (3.10+ required)
- Restart Jupyter kernel after installation

### Issue 5: TensorRT-LLM Build Failures

**Symptoms**: TensorRT-LLM compilation fails

**Solutions**:
- Ensure CUDA toolkit is installed (11.8 or 12.x)
- Verify GPU compute capability is supported (7.0+)
- Use pre-built containers from NGC when possible
- Check TensorRT-LLM documentation for build requirements
